[ガンマ関数](https://ja.wikipedia.org/wiki/%E3%82%AC%E3%83%B3%E3%83%9E%E9%96%A2%E6%95%B0)

$$
\begin{align}
    \Gamma \left( z\right) =\int _{0}^{\infty }t^{z-1}e^{-t}dt
\end{align}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

gamma = lambda t,z : t**(z-1)*np.exp(-t)

L = np.linspace(0,20,100)

plt.figure(facecolor="white", edgecolor="black", linewidth=2)
plt.plot(L,gamma(L,4))
plt.show()
```
