```python
import matplotlib.pyplot as plt
from matplotlib.patches import ArrowStyle
import numpy as np
```

```python
gamma = lambda t,z : t**(z-1)*np.exp(-t)
L = np.linspace(0,20,100)
```

```python
plt.figure(figsize=(10,6), dpi=80, facecolor="white", edgecolor="black", linewidth=2)

Y = gamma(L,4)

Ls, Ys = L/2+0.5,Y/Y.max()*0.7+1
Lsp = np.append(0,Ls)
Ysp = np.append(1,Ys)
plt.plot(Lsp, Ysp, color='blue')

Ls, Ys = L/2+1.5,Y/Y.max()*0.7+2
Lsp = np.append(0,Ls)
Ysp = np.append(2,Ys)
plt.plot(Lsp, Ysp, color='blue')

plt.grid()
# x軸を0〜7まで0.5刻みで表示
plt.xticks(np.arange(0, 5, 1),["", "$i-2$", "$i-1$", "$i$", "$i+1$"], fontsize=15)
# y軸を0〜8まで3刻みで表示
plt.yticks(range(0, 3, 1),["$n-1$", "$n$", "$n+1$"], fontsize=15)

plt.xlim(0,5)
plt.ylim(0,3)

plt.ylabel('time', fontsize=15)
plt.xlabel('space', fontsize=15)

plt.savefig('ref/wave01.svg', bbox_inches="tight" )

plt.show()
```
